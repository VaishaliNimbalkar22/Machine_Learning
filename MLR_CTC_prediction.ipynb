```python
# Reading data
df = pd.read_csv("/content/CTCdata - SLR MLR (2).csv")
df.head()
```

```python
# separating target and independent variables
y=df[["CTCoffered"]]
x=df[df.columns[1:]]
```

```python

#spliting the data into train & test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Model object
MLR=LinearRegression()
```

```python
#training model
MLR.fit(x_train,y_train)
```

```python
#predict
y_test['y_pred']=MLR.predict(x_test)
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test['CTCoffered'],y_test['y_pred'])
mse
```

```python
import math
```

```python
RMSE=math.sqrt(mse)
RMSE
```

```python
r_sq=MLR.score(x_train,y_train)
r_sq
```
